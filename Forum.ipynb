```python
import pandas as pd
import numpy as np
from sklearn import svm
from sklearn import metrics
from sklearn.metrics import accuracy_score
```

```python
internet = pd.read_csv("gapminder_internet.csv")
internet.dropna(inplace = True) #To erase all NAN value
print(internet)

mean = internet["internetuserate"].mean() #take the mean/average

internet["UserRate"] = np.where(internet["internetuserate"] > mean, "Often", "Rarely") #Make another column to show wether people in the country is often usen the inter or no
internet
```

```python
X=internet.iloc[0:209,[1,3]] #Slice the csv to take incomeperson and urbanrate
y=internet.loc[0:212,'UserRate'] #Slice to take userate

clf = svm.SVC(kernel='rbf', random_state=5, gamma=5, C=5)

clf.fit(X, y) #This is to train the data to predict (This fit needs to be called before you do predict)

y_pred=clf.predict(X)


print(accuracy_score(y, y_pred)) #Accuracy score from the predicted and previous
```
